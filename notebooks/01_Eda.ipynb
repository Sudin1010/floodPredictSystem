Step 10: Remove ID column for EDA

```python
df_eda = df.drop("id", axis=1)
df_eda.head()
```

Step 11: Target variable distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df_eda["FloodProbability"], kde=True)
plt.title("Distribution of Flood Probability")
plt.xlabel("Flood Probability")
plt.ylabel("Frequency")
plt.show()
```

Step 12: Feature distribution

```python
df_eda.drop("FloodProbability", axis=1).hist(figsize=(18,14), bins=20)
plt.suptitle("Feature Distribution of Original Dataset")
plt.show()
```

Step 13: Correlation heatmap

```python
plt.figure(figsize=(16,12))
corr = df_eda.corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Original Dataset")
plt.show()
```

Step 14: Correlation with target

```python
corr = df_eda.corr()
target_corr = corr["FloodProbability"].sort_values(ascending=False)

plt.figure(figsize=(8,10))
target_corr.drop("FloodProbability").plot(kind="barh")
plt.title("Feature Correlation with Flood Probability")
plt.xlabel("Correlation Value")
plt.ylabel("Features")
plt.show()
```
